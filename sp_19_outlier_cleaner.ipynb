```python
import numpy as np

# Define the columns and methods
columns = {
    'Rating': 'IQR',
    'Easy Apply': 'Z-Score',
    'Min Salary (K)': 'IQR',
    'Max Salary (K)': 'IQR',
    'Avg Salary (K)': 'IQR',
    'Company_Encoded': 'None',
    'Salary Range (K)': 'IQR',
    'Log Avg Salary': 'Z-Score'
}

# Define the thresholds
thresholds = {
    'Easy Apply': 2,
    'Log Avg Salary': 2
}

# Apply the methods
for column, method in columns.items():
    if method == 'IQR':
        Q1 = df_original[column].quantile(0.25)
        Q3 = df_original[column].quantile(0.75)
        IQR = Q3 - Q1
        df = df_original[~((df_original[column] < (Q1 - 1.5 * IQR)) | (df_original[column] > (Q3 + 1.5 * IQR)))]
    elif method == 'Z-Score':
        threshold = thresholds[column]
        z_scores = np.abs((df_original[column] - df_original[column].mean()) / df_original[column].std())
        df = df_original[z_scores < threshold]
    elif method == 'None':
        pass  # No outlier removal for this column

print("Original dataset shape:", df_original.shape)
print("Filtered dataset shape:", df.shape)
```

```python
columns = {
    'Rating': 'IQR',
    'Easy Apply': 'None',
    'Min Salary (K)': 'IQR',
    'Max Salary (K)': 'IQR',
    'Avg Salary (K)': 'IQR',
    'Company_Encoded': 'None',
    'Salary Range (K)': 'IQR',
    'Log Avg Salary': 'IQR'
}

# Apply the methods
for column, method in columns.items():
    if method == 'IQR':
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        df_filtered = df[~((df[column] < (Q1 - 2.5 * IQR)) | (df[column] > (Q3 + 2.5 * IQR)))]
    elif method == 'None':
        pass  # No outlier removal for this column

print("Original dataset shape:", df_original.shape)
print("Filtered dataset shape:", df_filtered.shape)
```

```python
df_filtered.describe()
```

```python
df_filtered['Company_Encoded'] = np.log(df_filtered['Company_Encoded'])
```

```python
df_filtered['Log Avg Salary'] = np.sqrt(df_filtered['Log Avg Salary'])
df_filtered['Max Salary (K)'] = np.log(df_filtered['Max Salary (K)'])
df_filtered['Min Salary (K)'] = np.log(df_filtered['Min Salary (K)'])
```

```python
df_filtered.describe()
```

```python
# Function to detect and convert data type
def convert_data_type(df_filtered):
    for column in df_filtered.columns:
        if df_filtered[column].dtype == 'object':
            if df_filtered[column].nunique() <= 10:  # Assume it's categorical if unique values are less than or equal to 10
                df_filtered[column] = df_filtered[column].astype('category')
            else:
                try:
                    df_filtered[column] = pd.to_numeric(df_filtered[column])
                except ValueError:
                    pass
        elif df_filtered[column].dtype == 'int64':
            if df_filtered[column].nunique() == 2 and df_filtered[column].min() == 0 and df_filtered[column].max() == 1:
                df_filtered[column] = df_filtered[column].astype('bool')
    return df_filtered

# Convert data type of all columns
df_filtered = convert_data_type(df_filtered)
```

```python
df=df_filtered.drop(columns=[	'Unnamed: 0.2'	,'Unnamed: 0.1'	,'Unnamed: 0'])
```
